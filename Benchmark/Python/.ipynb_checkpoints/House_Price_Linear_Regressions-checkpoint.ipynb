# Statsmodels : OLS

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
# Formula base
formula="Price ~ Income_Area + Age_House + Room_Numbers + Bedrooms_Numbers + Population"

formula_0=smf.ols(formula=formula , data=data).fit()
print(formula_0.summary())
```

```python
#Number of parameters
print(f"The number of parameters used in this Linear model are {len(formula_0.params)}")
print("---------------------------")
#Lets extracts some metrics I see in the summary
print(f"The R squared from the model is {formula_0.rsquared}")
print(f"The R squared  adjusted from the model is {formula_0.rsquared_adj}")
print("---------------------------")
#lets see the mean squared error form the model
MSE1 =  np.mean(formula_0.resid**2)
print(f"The mean squared error is {MSE1}")
p1 = len(formula_0.params) # number of regressors
n = len(data)
MSE_adj1  = (n/(n-p1))*MSE1
print(f"The mean squared error adjusted is {MSE_adj1}")
```

```python
ols=pd.DataFrame({"$No_coef$":[len(formula_0.params)],"$R_2$":[formula_0.rsquared], "$R2_adjusted$":[formula_0.rsquared_adj],
                 "$MSE$":[MSE1],"$MSE_adjusted$":[MSE_adj1]}, index=["Results"])
ols
```

* The basic model´s result shows a higher $R_2$, but the $MSE$ doesn´t look good, because the values from each variables are totally dispersed, which prove the model is not appropiate for this case, although the score is very high. In addition to the above I am using all the data so that 

# Preprocessing:

* Before building various linear models, its important to normalize the information of each column from dataset. The idea of this technique is based the models from Machine Learning (Supervised Learning) perform a good prediction and drop good metrics. if any researcher try to make a preprocessing step before building a Machine Learning Algorithm, which means the performance will be succesful in many cases. So let´s do it.

```python
normalize=preprocessing.normalize(data[["Income_Area","Age_House","Room_Numbers","Bedrooms_Numbers","Population","Price"]])
```

```python
names=["Income_Area","Age_House","Room_Numbers","Bedrooms_Numbers","Population","Price"]
scaled_df = pd.DataFrame(normalize, columns=names)
scaled_df.head()
```

```python
fig, ax=plt.subplots(2,3, figsize=(16,6))

scaled_df["Income_Area"].plot(kind="hist",bins=50, ax=ax[0,0], color="blue")
scaled_df["Age_House"].plot(kind="hist",bins=50, ax=ax[0,1], color="red")
scaled_df["Room_Numbers"].plot(kind="hist",bins=50, ax=ax[0,2], color="purple")
scaled_df["Bedrooms_Numbers"].plot(kind="hist",bins=50, ax=ax[1,0], color="yellow")
scaled_df["Population"].plot(kind="hist",bins=50, ax=ax[1,1], color="green")
scaled_df["Price"].plot(kind="hist",bins=50, ax=ax[1,2], color="brown")
plt.show()
```

# Linear Regression with Sklearn

```python
X=scaled_df.drop(columns="Price", axis=1)
Y=scaled_df.Price
print(X.shape)
print(Y.shape)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y, test_size=0.33, random_state=1234, shuffle=True)
```

```python
print(f"Los datos de entrenamiento de X train: {X_train.shape}")
print(f"Los datos de entrenamiento de X train: {X_test.shape}")
print(f"Los datos de entrenamiento de X train: {Y_train.shape}")
print(f"Los datos de entrenamiento de X train: {Y_test.shape}")
```

```python
#Important step to deploy the metrics of each linear model: Basic Model, Lasso model, Ridge model, ElasticNet models
ols_model={}
lasso_model={}
ridge_model={}
elasticnet_model={}

def metrics_sklearn(true, predict):
    '''
    In this section, I deploy the main metrics used in sklearn that are very useful for each Supervised model (ML):
    '''
    mae= metrics.mean_absolute_error(true, predict)
    mse=metrics.mean_squared_error(true, predict)
    rmse= np.sqrt(metrics.mean_squared_error(true,predict))
    R2_squared=metrics.r2_score(true, predict)
    print(f"MAE: {mae}")
    print(f"MSE: {mse}")
    print(f"RMSE: {rmse}")
    print(f"R2: {R2_squared}")
    
def evaluate(true, predicted):
    '''
    In this section, I deploy the main metrics used in sklearn that are very useful for each Supervised model (ML):
    '''
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    return [mae, mse, rmse, r2_square]
```

```python
model=LinearRegression(fit_intercept=True)
model.fit(X_train,Y_train)
```

```python
# predicting the models using only the values from the X_test
predict_ols=model.predict(X_test)
predict_ols[:10].round(3)
```

```python
predict_ols_1=model.predict(X_train)
print(predict_ols_1[:10].round(1))
```

```python
#testing the values using only the train data:
results_metrics_0=metrics_sklearn(Y_train, predict_ols_1)
```

```python
#testing the values using only the test data:
results_metrics=metrics_sklearn(Y_test, predict_ols)
```

```python
rmse_ols=mean_squared_error(Y_test,predict_ols, squared=True)
socre_ols=model.score(X_test,Y_test)
print(f"the mean squared error from OLS {rmse_ols}")
print(f"the score from OLS {socre_ols}")
```

```python
nuevo=evaluate(Y_test,predict_ols)

ols_model={"MAE":nuevo[0], "MSE":nuevo[1],"RMSE":nuevo[2],"$R_2$":nuevo[3]}
ols_model_result=pd.DataFrame(ols_model, index=["Results"])
ols_model_result
```

# Regularization technique: Lasso.

```python
lassocv=LassoCV(cv=5,
             alphas=np.logspace(-10,2,200),
               random_state=42)

lassocv.fit(X_train,Y_train)
```

```python
# I want to know the best alpha found on lassocv

print(f"The best alpha found on LASSOCV is {lassocv.alpha_}")
```

# Testing the lasso regression again:

```python
# The reason I use only the lasso module is I just want to focus in 
lasso=Lasso(alpha=lassocv.alpha_,
           random_state=42)
#fittinh the model
lasso.fit(X_train,Y_train)
```

```python
lasso_train=lasso.predict(X_train)
print(f"the values predicted using only the values from X_train: {lasso_train[:6]}")
lasso_test=lasso.predict(X_test)
print(f"the values predicted using only the values from X_test: {lasso_test[:6]}")
```

```python
result_metrics_1=metrics_sklearn(Y_train,lasso_train)
```

```python
result_metrics_1=metrics_sklearn(Y_test,lasso_test)
```

```python
lasso_result=pd.DataFrame({"Predictors": X_train.columns, "coef":lasso.coef_})
lasso_result
```

# Regularization technique: Ridge.

```python
ridgecv=RidgeCV(
                alphas=np.logspace(-10,1,100),
               store_cv_values=True,
               fit_intercept=True)

#fitting the ridge model:
ridgecv.fit(X_train,Y_train)
```

```python
# I want to know the best alpha found on lassocv

print(f"The best alpha found on RIDGECV is {ridgecv.alpha_}")
```

# Testing the Ridge model:

```python
alpha=ridgecv.alpha_

ridge=Ridge(alpha=alpha, fit_intercept=True, random_state=42)
ridge.fit(X_train,Y_train)
```

```python
ridge_train=ridge.predict(X_train)
print(f"the values predicted using only the values from X_train: {ridge_train[:6]}")
ridge_test=ridge.predict(X_test)
print(f"the values predicted using only the values from X_test: {ridge_test[:6]}")
```

```python
result_metrics_2=metrics_sklearn(Y_train,ridge_train)
```

```python
result_metrics_2=metrics_sklearn(Y_test,ridge_test)
```

```python
pd.Series(X_train.columns, ridge.coef_)
```

```python
coefs = []

alphas = np.logspace(-10,2, 200)

# Train the model with different regularisation strengths
for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(X_train, Y_train)
    coefs.append(ridge.coef_)

plt.figure(figsize=(20, 6))

plt.subplot(121)
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale("log")
plt.xlabel("alpha")
plt.ylabel("weights")
plt.title("Ridge coefficients as a function of the regularization")
plt.axis("tight")
```

# Elastic-Net Technique:

```python
elasticcv=ElasticNetCV(cv=5,
                    alphas=np.logspace(-10,1,100),
                    random_state=0)
#Fitting the model:
elasticcv.fit(X_train,Y_train)
```

```python
print(f"The best alpha found on Elastic-Net: {elasticcv.alpha_}")
```

```python
alpha=elasticcv.alpha_

elastic=ElasticNet(alpha=alpha)
#fitting the model:
elastic.fit(X_train,Y_train)
```

```python
print("The parameters found on elastic net are:", elastic.coef_)
```

```python
elastic_train=elastic.predict(X_train)
print(f"the values predicted using only the values from X_train: {elastic_train[:6]}")
elastic_test=elastic.predict(X_test)
print(f"the values predicted using only the values from X_test: {elastic_test[:6]}")
```
