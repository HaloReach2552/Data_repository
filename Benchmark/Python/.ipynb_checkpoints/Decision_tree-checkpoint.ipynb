```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics
```

```python
path = "https://raw.githubusercontent.com/HaloReach2552/Data_repository/main/Benchmark/data/bank.csv"

df = pd.read_csv(path)
df.head()
```

```python
def missing_values(data):
    count = data.isna().sum().any()
    if count == True:
        print("Ok, There are missing values in this data")
    else:
        print("Everything is ok....")

missing_values(df)
```

```python
def unique_values(data):
    columnas =['age', 'job', 'marital', 'education', 'default', 'balance', 'housing','loan', 
               'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 
               'poutcome', 'deposit']
    
    for elemento in columnas:
        resultado = np.unique(data[elemento])
        nr_values = len(resultado)
        if nr_values < 10:
            print(f"The column {elemento} from the data has {nr_values}")
        else:
            print(f"The column {elemento} from the data has {nr_values}")

unique_values(df)
```

```python
trabajos = ['admin.','technician' ,'services', 'management' ,'retired', 'blue-collar','unemployed', 'entrepreneur' ,
            'housemaid', 'unknown','self-employed','student']

for trabajo in trabajos:
    job = len(df[(df["job"]==trabajo) & (df["deposit"]=="yes")])
    print("El trabajo {} deposita {}".format(trabajo,job))
```

```python
martial = ['married', 'single', 'divorced']

for i in martial:
    estado = [df[(df["marital"] == i) & (df["deposit"] == "yes")].shape[0]]
    print(f"El estado : {i} de las personas que depositan es de {estado}")
```

```python
#Replacing the string values to numeric values 
df["job"] = df["job"].replace(['management', 'admin.'], 'white-collar')
df["job"] = df["job"].replace(['services','housemaid'], 'pink-collar')
df["job"] = df["job"].replace(['retired', 'student', 'unemployed', 'unknown'], 'other')
df['poutcome'] = df['poutcome'].replace(['other'] , 'unknown')

df["new_default"] = df["default"].replace({"no":0, "yes":1})
df["new_housing"] = df["housing"].replace({"yes":1,"no":0})
df["new_loan"] = df["loan"].replace({"yes":1,"no":0})
df["new_deposit"] = df["deposit"].replace({"yes":1, "no":0})

#Dropping some columns
df.drop("default", axis=1, inplace=True)
df.drop("housing", axis=1, inplace=True)
df.drop("loan", axis=1, inplace=True)
df.drop("deposit", axis=1, inplace=True)
df.drop("contact", axis=1, inplace=True)
df.drop("month", axis= 1, inplace=True)

#Changing the type of values from some columns
df["new_default"] = df["new_default"].astype("category")
df["new_housing"] = df["new_housing"].astype("category")
df["new_loan"] = df["new_loan"].astype("category")
df["new_deposit"] = df["new_deposit"].astype("category")

df.info()
```

```python
one_hot = pd.get_dummies(df, columns = ['job',"marital","education","poutcome"] , \
                                   prefix = ['job', 'marital', 'education', 'poutcome'])
one_hot.head(5)
```

```python
one_hot.drop("pdays", axis= 1 , inplace=True)
one_hot.head(5)
```

```python
plt.figure(figsize=(5,6))

plt.hist(one_hot["age"], color= "b", label= "age", bins = 20)
plt.xlabel("")
plt.ylabel("count", size = 10)
plt.axvline(np.mean(one_hot["age"]), color = "black", linewidth=3, label="mean")
plt.axvline(np.median(one_hot["age"]), color = "red", linewidth=3, label="median")
plt.title("Age´s Distribution", size =20)
plt.legend(loc="best")
plt.show()
```

# Building the model

```python
x = one_hot.drop("new_deposit", axis = 1)
y = one_hot.new_deposit

X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size= 0.33, random_state= 42)

def decision_tree(criterio,profundidad, aleatorio):
    dt2 = tree.DecisionTreeClassifier(criterion=criterio, max_depth=profundidad, random_state=aleatorio)
    dt2.fit(X_train, Y_train)
    dt2_score_train = dt2.score(X_train, Y_train)
    print("Training score: ",dt2_score_train)
    dt2_score_test = dt2.score(X_test, Y_test)
    print("Testing score: ",dt2_score_test)

    plt.figure(figsize=(16,6))
    plot_tree(decision_tree=dt2, feature_names=X_train.columns, filled=True, fontsize=10)
    plt.show()
```

```python
decision_tree("gini",5,42)
```

```python
decision_tree("entropy",4,42)
```
