```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://raw.githubusercontent.com/HaloReach2552/Data_repository/main/Benchmark/data/netflix_titles.csv"
data = pd.read_csv(url)
data.head(5)
```

```python
data.info()
```

```python
del data["show_id"]
del data["description"]
```

```python
data.head(6)
```

```python
data.columns = ["Tipo","Titulo","Director","Personajes","Pais","Mes","Año","Clasificacion","Duracion","Genero"]
data.head(5)
```

```python
data.dropna(axis=0, inplace=True)
data.info()
```

```python
data.head(4)
```

```python
Tipo = data["Tipo"].value_counts()
Tipo
```

```python
sns.countplot(data["Tipo"], dodge="Tipo")
plt.title("Movies vs Tv Shows")
plt.show()
```

```python
paises = data["Pais"].value_counts()[:15]
paises
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=paises.index, y=paises.values)
plt.xticks(rotation = 45)
plt.title("Best 14 countries in Netflix")
plt.show()
```

```python
clasificacion = data.Clasificacion.value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=clasificacion.index, y=clasificacion.values)
plt.xticks(rotation = 45)
plt.title("Types of clasification in Netflix according the movies")
plt.show()
```
